## Feature Engineering

This notebook prepares the integrated AltCredit dataset for machine learning
by encoding categorical variables, scaling numerical features, and creating
trainâ€“test splits. The objective is to produce model-ready feature matrices
while preserving interpretability and preventing data leakage.


```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/final_dataset.csv")
df.shape
```

```python
X = df.drop(columns=["TARGET", "SK_ID_CURR"])
y = df["TARGET"]
```

```python
cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(include=["int64", "float64"]).columns

len(cat_cols), len(num_cols)
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=cat_cols,
    drop_first=True
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
X_train.shape, X_test.shape
y_train.value_counts(normalize=True)
y_test.value_counts(normalize=True)
```

```python
X_train.to_csv("../data/processed/X_train.csv", index=False)
X_test.to_csv("../data/processed/X_test.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)
```
